```python
import os
import string
import json
import datetime
import pandas
import hatebase
import spacy
from spacy_langdetect import LanguageDetector

def load_channel(filename):
    channel_messages = []
    with open(''.join(['backup_january/', filename]), 'r') as f:
        channel_messages = f.read().split(';;;\n;;;')
    f.close()
    channel_messages = [eval(msg)for msg in channel_messages]
    return channel_messages
```

```python
text_files = [tf for tf in os.listdir('backup') if tf.endswith('txt')]
```
